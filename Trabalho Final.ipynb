```python
import pandas as pd
import random
```

```python
def jogo(jogador1, opcao1, jogador2, opcao2, tabela):

  #iniciando a variável (sem essa linha td quebra)
  ganhador = None

  for i in range(len(tabela)):

    if (opcao1 == 'Pedra' and opcao2 == 'Tesoura') or (opcao1 == 'Papel' and opcao2 == 'Pedra') or (opcao1 == 'Tesoura' and opcao2 == 'Papel'):
      ganhador = jogador1
    elif (opcao2 == 'Pedra' and opcao1 == 'Tesoura') or (opcao2 == 'Papel' and opcao1 == 'Pedra') or (opcao2 == 'Tesoura' and opcao1 == 'Papel'):
      ganhador = jogador2
    elif(jogador1 < jogador2):
      ganhador = jogador1
    else:
      ganhador = jogador2

    return ganhador
```

```python
tabela = pd.read_csv('campeonato.csv')

#remove coluna inútil
tabela = tabela.drop(columns=['Unnamed: 0'])
```

#### Primeira Rodada

```python
#sorteia a ordem dos jogadores da rodada
ordem = tabela.columns.tolist()
random.shuffle(ordem)
print(ordem)
```

```python
#criação da lista colunas
colunas = ['Jogador 1', 'Jogador 2', 'Resultado']

#usa os itens da lista para fazer as colunas do dataframe campeonato
campeonato = pd.DataFrame(columns=colunas)
```

```python
def jogadores():

  partida = 0

  #escreve o nome do jogador 1 e 2 de cada dupla nas tuplas
  for i in range(0, len(ordem) -1, 2):
    campeonato.loc[partida, 'Jogador 1'] = ordem[i]
    campeonato.loc[partida, 'Jogador 2'] = ordem[i+1]
    partida +=1
```

```python
def classifica(campeonato, ordem, jogo, tabela):

  ganhadores = []
  jogadores()

  for i in range(len(campeonato)):

    #define os jogadores
    jogador1 = campeonato.loc[i, campeonato.columns[0]]
    jogador2 = campeonato.loc[i, campeonato.columns[1]]

    #localiza as jogadas definidas na tabela campeonato.csv
    opcao1 = tabela.loc[i, jogador1]
    opcao2 = tabela.loc[i, jogador2]

    #roda a função jogo e completa a lista jogadores
    ganhador = jogo(jogador1, opcao1, jogador2, opcao2, tabela)
    campeonato.loc[i, 'Resultado'] = ganhador
    ganhadores.append(ganhador)

  #refaz a tabela apenas com ganhadores
  tabela = tabela[tabela.columns.intersection(ganhadores)]

  return tabela
```

```python
#chama função classifica
tabela = classifica(campeonato, ordem, jogo, tabela)
```
